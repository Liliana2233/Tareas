**P6.** Crea los siguientes arreglos e imprime su `shape` y `dtype`:  
- Un arreglo con los valores `[1.5, 2.5, 3.5, 4.5]`  
- Una matriz de ceros de 3 filas × 4 columnas  
- Un arreglo con 6 valores igualmente espaciados entre 0 y 1 (usando `linspace`)

```python

import numpy as np

# 1. Arreglo con valores decimales
arreglo = np.array([1.5, 2.5, 3.5, 4.5])
print("Arreglo:", arreglo)
print("Shape:", arreglo.shape)
print("Dtype:", arreglo.dtype)
print()

# 2. Matriz de ceros 3x4
matriz_ceros = np.zeros((3, 4))
print("Matriz de ceros:")
print(matriz_ceros)
print("Shape:", matriz_ceros.shape)
print("Dtype:", matriz_ceros.dtype)
print()

# 3. Arreglo con linspace
arreglo_linspace = np.linspace(0, 1, 6)
print("Arreglo linspace:", arreglo_linspace)
print("Shape:", arreglo_linspace.shape)
print("Dtype:", arreglo_linspace.dtype)
```

**P7.** Dado el arreglo `a = np.arange(1, 13)`, usa `reshape` para convertirlo en una matriz de 3 × 4.  
Luego imprime el valor máximo, el valor mínimo y la posición (índice) del máximo.

```python

import numpy as np

# Crear el arreglo y convertirlo en una matriz 3x4
a = np.arange(1, 13)
matriz = a.reshape(3, 4)

print("=== Matriz 3 x 4 ===")
print(matriz)

# Calcular valores
valor_maximo = matriz.max()
valor_minimo = matriz.min()
indice_maximo = tuple(int(x) for x in np.unravel_index(matriz.argmax(), matriz.shape))

print("\n=== Resultados ===")
print(f"Valor máximo: {valor_maximo}")
print(f"Valor mínimo: {valor_minimo}")
print(f"Posición del valor máximo (fila, columna): {indice_maximo}")

```

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,  
usa **indexación booleana** para obtener solo las notas mayores o iguales a 6. Imprime el resultado.

```python

import numpy as np

# Arreglo de notas
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

# Indexación booleana: notas mayores o iguales a 6
notas_aprobadas = notas[notas >= 6]

print("=== Notas Aprobadas (>= 6) ===")
print(notas_aprobadas)

```
