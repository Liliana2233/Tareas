**En este dataset las variables `Id` y `Year` representan variables categorica, convertir estas columnas a su formato correcto**

```python
# Se convierten las columnas Id y Year a tipo categoria utilizando astype("category"), estos representan
#identificadores y años respectivamente y no valores numericos para calculo

salarios["Id"] = salarios["Id"].astype("category")
salarios["Year"] = salarios["Year"].astype("category")

salarios.info()
```

**la variable `Year` es categorica ordinal, primero detecte cuales son los valores unicos que tiene la columna**

```python
# Se utiliza el método unique() para identificar los valores únicos de la columna Year.
# Se observa que los años presentes son 2011, 2012, 2013 y 2014,
# lo que confirma que es una variable categórica ordinal.

salarios["Year"].unique()
```

**Apartir de la información anterior convierta la columna `Year` en variable categorica ordinal y visualice la columna**

```python
# Se visualizan los primeros valores de la columna Year, donde se observa que
# los registros iniciales corresponden al año 2011.

salarios["Year"] = pd.Categorical(
    salarios["Year"],
    categories=[2011, 2012, 2013, 2014],
    ordered=True
)

salarios["Year"].head()
```

**Realice nuevamente una descripcion estadistica de las columnas numericas?**

**¿Cuál es el título de trabajo de JOSEPH DRISCOLL? Nota: Utilice mayúsculas, de lo contrario, puede obtener una respuesta que no coincida.**

```python
#Se filtra el DataFrame utilizando el nombre del empleado JOSEPH DRISCOLL
#El resultado muestra que su título es CAPTAIN, FIRE SUPPRESSION.

salarios[salarios["EmployeeName"] == "JOSEPH DRISCOLL"]["JobTitle"]
```

**¿Cuanto gana JOSEPH DRISCOLL (incluyendo beneficios)?**

```python
# Se Se busca el empleado JOSEPH DRISCOLL en el DataFrame y se extrae su salario
# Total con beneficios, obteniendo un valor de 270324.91


salarios[salarios["EmployeeName"] == "JOSEPH DRISCOLL"]["TotalPayBenefits"].iloc[0]
```

**¿Cual es el nombre de la persona con mejor pago (incluyendo beneficios)?**

```python
# Se busca el valor máximo de la columna TotalPayBenefits y luego se identifica
# Que empleado que tiene mejor pago es NATHANIEL FORD


salarios[salarios["TotalPayBenefits"] == salarios ["TotalPayBenefits"].max()]["EmployeeName"]
```

**¿Cual es el nombre de la persona peor paga (incluyendo beneficios)? ¿Notan algo extraño sobre cuánto le pagan?**

```python
# Se identifica la persona con el menor valor en la columna TotalPayBenefits,
# obteniendo que es JOE LOPEZ.

salarios.loc[salarios["TotalPayBenefits"].idxmin(), "EmployeeName"]
```

**Cuantos registros tienen un valor de TotalPayBeneficts menor que cero**

**¿Cual es el promedio (mean) BasePay de todos los empleados por año? los años van entre (2011-2014)?** (Tip: Usar groupby)

```python
# Se agrupan los datos por año y se calcula el promedio del salario base
# para cada uno de los años

salarios.groupby("Year")["BasePay"].mean()
```

**¿Cuántos puestos de trabajo Unicos existieron en 2013? (Ejemplo, los títulos de trabajo con solo una ocurrencia en 2013?)**

```python
# Se seleccionan los datos del 2013 y se cuentan los trabajos que aparecen
# Solo una vez, obteniendo el total de puestos únicos.

salarios_2013 = salarios[salarios["Year"] == 2013]

(salarios_2013["JobTitle"].value_counts() == 1).sum()
```

**¿Cuántas personas tienen la palabra Chief en el título de su trabajo?** Apoyese en los métodos para manipular strings y en el método `.apply()`

```python
# Se utiliza el método apply() junto con una función lambda para verificar
# Si la palabra "Chief" está presente en cada título de trabajo

salarios["JobTitle"].apply(lambda x: "Chief" in x).sum()
```

Este código es valido, se recomienda hacer el código mas simple y agregar typing y documentación.

**Bono: ¿Existe una correlación entre la longitud del nombre del Título del trabajo y el Salario?**

crea una nueva columna donde se guarde la longitud del titulo del trabajo. tip: usar el método `.apply()`

```python
# Se calcula la longitud del título del trabajo y se compara con el salario.
# El resultado muestra que no hay relación entre ambas variable

salarios["JobTitle_length"] = salarios["JobTitle"].apply(len)

```

Utilice el método `.corr()` de Pandas para calcular la correlacion

```python
# La correlación es muy baja, por lo que no hay relación entre el tamaño del
# título del trabajo y el salario.

salarios[["JobTitle_length", "TotalPayBenefits"]].corr()
```
